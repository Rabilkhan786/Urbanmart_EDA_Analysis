```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

um = pd.read_csv("Data/UrbanMart_Transactions.csv")
```

```python
um.head(1)
um.tail(2)
um.info()
```

```python
um.shape
```

```python
#Checking how many columns are:
um.columns
```

```python
#Checking Data Types and change date datatype:
um.dtypes
um['TransactionDate']=pd.to_datetime(um['TransactionDate'])
um['TransactionValue']=pd.to_numeric(um['TransactionValue'])
```

```python
#Checking null value , duplicate value and replace it :
um.isna().sum() # NO null value 
```

```python
#Duplicate checking:
um.duplicated().sum() # NO Duplicated rows 
```

```python
# statistic Describe:
um.describe()
```

```python
#Transforming columns:
um['TransactionValue_Lakhs'] = um['TransactionValue']/100000
um['TransactionValue_Lakhs'] = um['TransactionValue_Lakhs'].round(2)
um['TransactionValue_Lakhs']
```

```python
# Checking  realted data  or different data from  the data set :
um['CustomerAgeGroup'].unique()
um['PaymentMethod'].unique()
um['Region'].unique()
um['ProductCategory'].unique()
um['TransactionDate'].unique()
um['TransactionID'].unique()
```

```python
# Transform Date column:
# um.drop(columns='Transcation_month',axis=1,inplace=True)
um['T_Month'] = um['TransactionDate'].dt.month
um['T_Month_Name'] = um['TransactionDate'].dt.month_name()
um['T_Year'] = um['TransactionDate'].dt.year
```

```python
# Removing unwanted spaces:
um['CustomerGender'] = um['CustomerGender'].str.strip()
um['CustomerID'] = um['CustomerID'].str.strip()
um['ProductCategory'] = um['ProductCategory'].str.strip()
um['PaymentMethod'] = um['PaymentMethod'].str.strip()
um['Region'] = um['Region'].str.strip()
um.head()
```

```python
#Renaming column name and lower_case:
um.rename(columns={
  'CustomerGender':'Gender',
  'CustomerAgeGroup' : 'Age_Group',
  'TransactionID' : 'Transaction_ID',
  'CustomerID' : 'Customer_ID',
  'TransactionDate': 'Transaction_Date',
  'ProductCategory' :'Product_Category',
  'TransactionValue' : 'Transaction_Value',
  'PaymentMethod' :'payment_Method'

},inplace=True)
um.columns = um.columns.str.lower()
um.head()
```

```python
#Save the cleaned data :
um.to_csv("Urbanmart_cleaned.csv",index= False)
```
